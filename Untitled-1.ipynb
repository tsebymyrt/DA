```python
# 카테고리별 주문 수 계산
category_cnt = product['Product_category_name'].value_counts().head(10)

plt.figure(figsize=(12, 8))
bars = category_cnt.sort_values(ascending=True).plot(kind='barh', color='skyblue')

plt.title('Top 10 제품 카테고리')
plt.xlabel('주문 수')
plt.ylabel('제품 카테고리')

# 각 바 위에 값 표시
for bar in bars.containers[0]:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2,
             f'{int(bar.get_width())}', va='center')

plt.show()
```

```python
# 상위 10개 도시 선택
top10_cities = customer['Customer_city'].value_counts().head(10).index

# 상위 10개 도시에 대한 데이터만 필터링
data = customer[customer['Customer_city'].isin(top10_cities)]

# 시각화
plt.figure(figsize=(8,6))
ax = sns.countplot(x='Customer_city', data=data,
                   order=data['Customer_city'].value_counts().index, 
                   palette='icefire_r')

plt.title('Top 10 도시명')
plt.xticks(rotation=45) # 도시 이름이 긴 경우를 위해 x축 라벨 회전
```

```python
# 지불 형태별 개수 계산
category_counts = payments['Payment_type'].value_counts()

# 바 그래프 그리기
fig, ax = plt.subplots(figsize=(12, 7))
bars = ax.bar(category_counts.index, category_counts)

# 각 바에 카운트 값 표시
for bar in bars:
    height = bar.get_height()
    ax.annotate(f'{height}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),  # 3 points vertical offset
                textcoords="offset points",
                ha='center', va='bottom')

ax.set_xlabel('지불형태')
ax.set_ylabel('갯수')
ax.set_title('지불형태 종류')

plt.show()
```

KPI Definition -1 월별매출 시각화

```python
#원본 데이터 유지를 위해 복사하여 사용
order = orders.copy()

order['Order_purchase_timestamp'] = pd.to_datetime(order['Order_purchase_timestamp'])
order['InvoiceYearMonth'] = order['Order_purchase_timestamp'].dt.strftime('%Y%m')

merged_orders = pd.merge(order, order_items, on='Order_id', how='left')
merged_orders['Revenue'] = merged_orders['Price'] * merged_orders['Order_item_id']
monthly_revenue = merged_orders.groupby('InvoiceYearMonth')['Revenue'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='InvoiceYearMonth', y='Revenue', data=monthly_revenue, marker='o')
plt.title('월별 매출 추이')
plt.xlabel('년월')
plt.ylabel('매출')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
orders_customers = pd.merge(order, customer, on='Customer_id', how='inner')
orders_monthly_active = orders_customers.groupby('InvoiceYearMonth')['Customer_unique_id'].nunique().reset_index()

orders_monthly_active.head()
```

```python
# 첫 구매인 주문 식별하기
# 주문 시간을 기준으로 첫 주문만 필터링
first_orders = orders.sort_values(by='Order_purchase_timestamp').drop_duplicates(subset='Customer_id', keep='first')
#결합
first_order_items = pd.merge(first_orders, order_items, on='Order_id')
first_order_items = pd.merge(first_order_items, product, on='Product_id')
# 첫 구매가 많은 제품 상위 10개
#top_10_products = first_order_items['Product_id'].value_counts().head(30)
#top_10_products_df = pd.DataFrame({'Product_id': top_10_products.index, 'Counts': top_10_products.values})
# 첫 구매가 많은 제품 카테고리 상위 10개
top_10_categories = first_order_items['Product_category_name'].value_counts().head(30)
top_10_categories_df = pd.DataFrame({'Product_category_name': top_10_categories.index, 'Counts': top_10_categories.values})

plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_categories_df, x='Product_category_name', y='Counts')
plt.title('첫 구매가 많은 상위 10개 제품 카테고리')
plt.xticks(rotation=45)
plt.show()

```

```python
order_customer = pd.merge(orders, customer, on='Customer_id')
order_item_product = pd.merge(order_items, product, on='Product_id')
full_data = pd.merge(order_customer, order_item_product, on='Order_id')
# 각 고객별 각 제품 카테고리 재구매 횟수 계산
repurchase_counts = full_data.groupby(['Customer_unique_id', 'Product_category_name']).size().reset_index(name='counts')
# 재구매 횟수가 2 이상인 데이터 필터
repurchases = repurchase_counts[repurchase_counts['counts'] > 1]
# 제품 카테고리별로 재구매 횟수 합계 계산
category_repurchase_counts = repurchases.groupby('Product_category_name')['counts'].sum().reset_index()
# 재구매 횟수가 많은 상위 10개 카테고리 추출
top_categories = category_repurchase_counts.sort_values(by='counts', ascending=False).head(30)
# 결과 출력
print(top_categories)
print(top_10_categories)
```
